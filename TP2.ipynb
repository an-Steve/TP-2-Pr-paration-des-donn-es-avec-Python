# **Exercice 5 â€” Traitement des outliers**

**1.Identifiez les valeurs nÃ©gatives dans Total Deaths**

```python
neg_total_deaths = df[df['Total Deaths'] < 0]
print("Valeurs nÃ©gatives dans Total Deaths :")
print(neg_total_deaths, "\n")
```

**2. Identifiez les valeurs aberrantes dans Magnitude (> 10)**

```python
outliers_magnitude = df[df['Magnitude'] > 10]
print("Valeurs aberrantes dans Magnitude (>10) :")
print(outliers_magnitude, "\n")
```

**3. DÃ©cidez : supprimer ou corriger ?**

- Total Deaths : valeurs nÃ©gatives impossible â†’ supprimer ces ligne
- Magnitude : valeurs >10 probablement erreurs â†’ capper Ã  10


**4.Appliquez le traitement**

```python
df = df[df['Total Deaths'] >= 0]            # Supprimer lignes avec Total Deaths nÃ©gatif
df['Magnitude'] = df['Magnitude'].clip(upper=10)  # Capper Magnitude Ã  10
```

**5. VÃ©rifiez le rÃ©sultat**

```python
print("VÃ©rification aprÃ¨s traitement :")
print("Nombre de valeurs nÃ©gatives dans Total Deaths :", (df['Total Deaths'] < 0).sum())
print("Nombre de valeurs supÃ©rieures Ã  10 dans Magnitude :", (df['Magnitude'] > 10).sum())
```
